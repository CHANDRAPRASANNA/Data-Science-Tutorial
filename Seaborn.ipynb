```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
#accessing tips dataset already present in seaborn
df=sb.load_dataset("tips")
print(df.to_string())
#FacetGrid work as subplot
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")#without hue we will get the same color for both smoking and non smoking persons
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#statistical estimation
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
sb.barplot(x="sex",y="tip",data=df)#without hue
plt.show()
```

```python
#linear Relationship
sb.regplot(x="total_bill",y="tip",data=df)#regression plot
plt.show()
```

```python
#pair plot is the combination of uni varient and bi varient
#diagonal elments are histograms are univarient
#other are scatter plot are bi varient
sb.pairplot(df)
plt.show()
```

```python
#pairplot seaborn:ploting selected variables
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
#Adding aHue color to a Seaborn pairplot
sb.pairplot(df,hue="size")
plt.show()
```

```python
#adding different colors
sb.pairplot(df,hue="size",palette="husl")#palette is for adding different colors for
plt.show()
```

```python
#Diagonal kind of plots
sb.pairplot(df,diag_kind="kde")#kde:kernal distributed estimator
plt.show()
```

```python
#Adjesting plot kind
sb.pairplot(df,diag_kind="reg")
plt.show()
```

```python
sb.paiplot(df,hue="sex",markers=['o','s'])
plt.show()
```

```python
g=sb.pairplot(df,hue="day")
g.fig.subtitle("pairplot of tips dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
